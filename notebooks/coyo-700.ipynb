```python
from phrasely.data_loading.coyo_loader import COYOLoader

loader = COYOLoader(
    batch_size=50_000,
    max_phrases=1_000_000,
    language_filter=None,  # or "en" if you want English only
)

phrases = []

for df in loader.stream_load():
    phrases.extend(df["phrase"].tolist())

len(phrases)
```

```python
import os

cache_dir = "data_cache/coyo_1m"
os.makedirs(cache_dir, exist_ok=True)
```

```python
import pandas as pd

shard_size = 50_000
total = len(phrases)

for i in range(0, total, shard_size):
    shard_phrases = phrases[i : i + shard_size]
    df = pd.DataFrame({"phrase": shard_phrases})

    shard_id = i // shard_size
    out_path = f"{cache_dir}/shard_{shard_id:04d}.parquet"

    print(f"Writing {out_path} ({len(df)} rows)")
    df.to_parquet(out_path, index=False)
```
